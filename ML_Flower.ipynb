```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

with open('iris.data','r') as file:
  dataset=pd.read_csv(file,sep=',')
d1=pd.DataFrame(dataset)
d1
import seaborn as sns
g = sns.pairplot(dataset,hue="Species")
dataset['Species']=dataset['Species'].replace(['Iris-setosa','Iris-versicolor','Iris-virginica'],['1','2','3'])

X=dataset.drop(columns=['Species'])
y=dataset['Species'].to_list()
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)
from sklearn.neighbors import KNeighborsClassifier
model1= KNeighborsClassifier(n_neighbors=3)
model1.fit(X_train,y_train)
model1.predict(X_test)
a_s=model1.score(X_test,y_test)
print(a_s*100,'%')
```
